```python
'''
把 STPS 数据读进来.
'''
import numpy as np

# 读取 dissimilarity_spaced.npy 文件
dissimilarity_spaced = np.load('../analysis/nps/dissimilarity_spaced.npy') # ..表示回退一级目录
dissimilarity_spaced = np.delete(dissimilarity_spaced, 24, axis=0)
# 读取 dissimilarity_massed.npy 文件
dissimilarity_massed = np.load('../analysis/nps/dissimilarity_massed.npy')
dissimilarity_massed = np.delete(dissimilarity_massed, 24, axis=0)
# 已经去掉prex048 第25被试

'''

找出cluster的时间点.

'''

import numpy as np
from scipy.stats import ttest_rel

# 计算t值和p值
t_values, p_values = ttest_rel(dissimilarity_spaced, dissimilarity_massed)

# 设置显著性水平
alpha = 0.05

# 找出差异显著的时间点
significant_timepoints = np.where(p_values < alpha)

# 寻找连续显著的cluster
clusters = []
current_cluster = []
previous_timepoint = None

for timepoint in significant_timepoints[0]:
    if not current_cluster:
        current_cluster.append(timepoint)
    elif timepoint == previous_timepoint + 1:
        current_cluster.append(timepoint)
    else:
        clusters.append(current_cluster)
        current_cluster = [timepoint]
    previous_timepoint = timepoint

if current_cluster:
    clusters.append(current_cluster)

print('显著差异的时间点：', significant_timepoints)
print('连续显著的cluster：', clusters)


'''
为了从dissimilarity_spaced和dissimilarity_massed中提取显著差异的部分，我们可以使用numpy的数组索引。假设你已经获得了significant_timepoints，可以按照以下方式提取相关部分：
'''

# 提取显著差异的部分
dissimilarity_spaced_significant = dissimilarity_spaced[:, significant_timepoints]
dissimilarity_massed_significant = dissimilarity_massed[:, significant_timepoints]

print('显著差异部分 - dissimilarity_spaced: \n', dissimilarity_spaced_significant)
print('显著差异部分 - dissimilarity_massed: \n', dissimilarity_massed_significant)

'''
dissimilarity_spaced_significant和dissimilarity_massed_significant的数据结构为ndarray(28,1,450),我想通过对时间维度进行平均,压缩到28,这样就是28个被试对应的均值,然后把这28个值合并到dataframe类型的数据filtered_pivoted_data里面,作为独立的一列.
'''

import numpy as np
import pandas as pd


# 沿时间维度进行平均，压缩到形状为(28,)
spaced_mean = np.mean(dissimilarity_spaced_significant, axis=-1).squeeze()
massed_mean = np.mean(dissimilarity_massed_significant, axis=-1).squeeze()

# 将平均值添加为新列
filtered_pivoted_data['spaced_mean_TPS'] = spaced_mean
filtered_pivoted_data['massed_mean_TPS'] = massed_mean

print(filtered_pivoted_data)
```

```python

'''

最简单的中介模型


'''
import numpy as np
import pandas as pd
import statsmodels.api as sm
import statsmodels.formula.api as smf

# 自变量应该是学习方法 0 1 来代表学习方法
# 这种做法确实就不如被试内精细.
zeros = np.zeros(28)
ones = np.ones(28)
# 拼接数组
xuexifangfa = np.concatenate((zeros, ones))
X = xuexifangfa

# 假设你已经有一个名为df_groupSub_effect_size的DataFrame，其中包含'effect_size_S2O'和'effect_size_M2O'两列
# 从df_groupSub_effect_size中提取'effect_size_S2O'和'effect_size_M2O'列，并将它们拼接为一个列的ndarray
M1 = np.concatenate((df_groupSub_effect_size['effect_size_S2O'].values, df_groupSub_effect_size['effect_size_M2O'].values))
# 用deltaLG来作为中介变量
M2 = np.concatenate((df_groupSub_average_deltameanLG['average_deltameanLG_so'].values, df_groupSub_average_deltameanLG['average_deltameanLG_mo'].values))
# 用average_deltaLG_s来作为中介变量
df_groupSub.loc[df_groupSub['sub'] == 'x016', :] = 0 #x16数据有问题,先补位0
M3 = np.concatenate((df_groupSub['average_deltaLG_s'].values, df_groupSub['average_deltaLG_m'].values))
# 带绝对值的情况
M4 = np.concatenate((df_groupSub['average_deltaLG_absm'].values, df_groupSub['average_deltaLG_abss'].values))
M = M1

# 用recognition来作为Y
Y = np.concatenate((df_allSub_finalTest['S'].values, df_allSub_finalTest['M'].values))
# 用 finaltest的logical score来作为Y
Y = np.concatenate((df_allSub_finalTest_LG['S'].values, df_allSub_finalTest['M'].values))


# 创建一个示例数据集，其中X为二元变量
np.random.seed(42)
# X = np.random.choice([0, 1], 100)
# M = 0.5 * X + np.random.normal(0, 1, 100)
# Y = 0.7 * M + np.random.normal(0, 1, 100)

data = pd.DataFrame({'X': X, 'M': M, 'Y': Y})

# 独立变量（X）与中介变量（M）的关系（a路径）
model_a = smf.ols("M ~ X", data=data).fit()

# 中介变量（M）与因变量（Y）的关系（b路径），同时控制独立变量（X）
model_b = smf.ols("Y ~ M + X", data=data).fit()

# 计算中介效应（a * b）和直接效应（c'路径）
indirect_effect = model_a.params["X"] * model_b.params["M"]
direct_effect = model_b.params["X"]

# 输出结果
print("Indirect effect (a * b):", indirect_effect)
print("Direct effect (c'):", direct_effect)
print("Total effect (c = a * b + c'):", indirect_effect + direct_effect)

# 输出a路径的估计值及其p值
print("a path estimate:", model_a.params["X"])
print("a path p-value:", model_a.pvalues["X"])

# 输出b路径的估计值及其p值
print("b path estimate:", model_b.params["M"])
print("b path p-value:", model_b.pvalues["M"])

# 输出c'路径的估计值及其p值
print("c' path estimate:", model_b.params["X"])
print("c' path p-value:", model_b.pvalues["X"])
```
